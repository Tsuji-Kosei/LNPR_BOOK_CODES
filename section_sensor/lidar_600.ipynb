```python
import pandas as pd       ###hist_600###
import matplotlib.pyplot as plt

data  = pd.read_csv("sensor_data_600.txt", delimiter=" ",
                    header=None, names = ("date","time","ir","lidar"))

data["lidar"].hist(bins = max(data["lidar"]) - min(data["lidar"]),align='left')
plt.show()
data
```

```python
data.lidar.plot() ###plot_all_data##
plt.show()
```

```python
data.ir.plot()
plt.show()
```

```python
data["hour"] = [e//10000 for e in data.time]  ###hourly_mean###
d = data.groupby("hour")
d.lidar.mean().plot()
plt.show()
```

```python
d.lidar.get_group(6).hist()     ###two_mode_hist###
d.lidar.get_group(14).hist()
plt.show()
```

```python
each_hour = { i : d.lidar.get_group(i).value_counts().sort_index()  for i in range(24)} #æ™‚é–“ã”ã¨ã«ãƒ‡ãƒ¼ã‚¿ãƒ•ãƒ¬ãƒ¼ãƒ ã‚’ä½œæˆ  ###calc_joint_probs
freqs = pd.concat(each_hour, axis=1) #concatã§é€£çµ
freqs = freqs.fillna(0)     #æ¬ æå€¤(NaN)ã‚’0ã§åŸ‹ã‚ã‚‹
probs = freqs/len(data) #é »åº¦ã‚’ç¢ºçŽ‡ã§

probs #è¡¨ç¤º
```

```python
import seaborn as sns   ###2d_hist ï¼ˆä¸‹ã®ã‚»ãƒ«ã‚‚ï¼‰

sns.heatmap(probs)
plt.show()
```

```python
sns.jointplot(data = data, x = data["hour"],y = data["lidar"], kind="kde")
plt.show()
```

```python
p_t = pd.DataFrame(probs.sum())   #å„åˆ—ã‚’åˆè¨ˆ   
p_t.plot()
p_t.transpose() #ç´™é¢ã®é–¢ä¿‚ã§è¡¨ã‚’æ¨ªä¸¦ã³ã«
```

```python
p_t.sum()    # 1ã«ãªã‚‹
```

```python
p_z = pd.DataFrame(probs.transpose().sum())        #è¡Œã¨åˆ—ã‚’è»¢ç½®ã—ã¦å„åˆ—ã‚’åˆè¨ˆ
p_z.plot()
p_z.transpose()
```

```python
p_z.sum()
```

```python
cond_z_t = probs/p_t[0]  #åˆ—ï¼ˆæ™‚é–“ï¼‰ã”ã¨ã«P(t)ã§å‰²ã‚‹ã¨P(x|t)ã¨ãªã‚‹   ###lidar600cond
cond_z_t
```

```python
(cond_z_t[6]).plot.bar(color="blue", alpha=0.5)  ###lidar600pxt###
(cond_z_t[14]).plot.bar(color="orange", alpha=0.5) 
plt.show()
```

```python
cond_t_z = probs.transpose()/probs.transpose().sum() #è¡Œã¨åˆ—ã‚’å…¥ã‚Œæ›¿ãˆã¦åŒæ§˜ã«è¨ˆç®—ã™ã‚‹ã¨P(t|z)ã¨ãªã‚‹  ###lidar600bayes1

print("P(z=630) = ", p_z[0][630]) #ã‚»ãƒ³ã‚µå€¤ãŒ630ã«ãªã‚‹ç¢ºçŽ‡ï¼ˆä½•æ™‚ã‹ã®æƒ…å ±ã¯ãªã„ï¼‰
print("P(t=13) = ", p_t[0][13]) #æ™‚é–“ãŒ13æ™‚ã§ã‚ã‚‹ç¢ºçŽ‡
print("P(t=13 | z = 630) = ", cond_t_z[630][13])
print("Bayes P(z=630 | t = 13) = ", cond_t_z[630][13]*p_z[0][630]/p_t[0][13])

print("answer P(z=630 | t = 13) = ", cond_z_t[13][630]) #13æ™‚ã«ã‚»ãƒ³ã‚µå€¤ãŒ630
```

```python
def bayes_estimation(sensor_value, current_estimation):  ###lidar600bayes2
    new_estimation = []
    for i in range(24):
        new_estimation.append(cond_z_t[i][sensor_value]*current_estimation[i])
        
    return new_estimation/sum(new_estimation) #æ­£è¦åŒ–
```

```python
estimation = bayes_estimation(630, p_t[0])  ###lidar600bayesonestep
plt.plot(estimation)
```

```python
values_5 = [630,632,636] #sensor_data_600.txtã‹ã‚‰æ‹¾ã£ã¦ããŸ5æ™‚å°ã®ã‚»ãƒ³ã‚µå€¤        ###lidar600bayesestm1

estimation = p_t[0]
for v in values_5:
    estimation = bayes_estimation(v, estimation)
    
plt.plot(estimation)
```

```python
values_11 = [617,624,619] #sensor_data_600.txtã‹ã‚‰æ‹¾ã£ã¦ããŸ11æ™‚å°ã®ã‚»ãƒ³ã‚µå€¤        ###lidar600bayesestm2

estimation = p_t[0]
for v in values_11:
    estimation = bayes_estimation(v, estimation)
    
plt.plot(estimation)
```
